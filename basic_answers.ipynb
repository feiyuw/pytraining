```python
# 1. get the most 5 active users from access_10000.log, and print their ip address and access count
# solution - 1: use dictionary
print 'this is solution - 1'
visitors = {}
with open('access_10000.log') as fp:
    for line in fp:
        ip = line.split(' ', 1)[0]
        if ip in visitors:
            visitors[ip] += 1
        else:
            visitors[ip] = 1
top_5_active_users = sorted(visitors.items(), key=lambda (ip, count): count, reverse=True)[:5]
print '\n'.join(['{}\t{}'.format(ip, count) for (ip, count) in top_5_active_users])

# solution - 2: use set and dictionary
print '\n\nthis is solution - 2'
with open('access_10000.log') as fp:
    visits = [e.split(' ', 1)[0] for e in fp]
visit_counts = {(e, visits.count(e)) for e in visits}
top_5_active_users = sorted(visit_counts, key=lambda (ip, count): count, reverse=True)[:5]
print '\n'.join(['{}\t{}'.format(ip, count) for (ip, count) in top_5_active_users])


# solution - 3: with itertools
print '\n\nthis is solution - 3'
import itertools
with open('access_10000.log') as fp:
    visits = (e.split(' ', 1)[0] for e in fp)
    visit_counts = [(ip, len(list(ips))) for (ip, ips) in itertools.groupby(sorted(visits))]
print '\n'.join(['{}\t{}'.format(ip, count) for (ip, count) in sorted(visit_counts, key=lambda (ip, count): count, reverse=True)[:5]])

# solution - 4: with collections
print '\n\nthis is solution - 4'
from collections import Counter
with open('access_10000.log') as fp:
    visits = (e.split(' ', 1)[0] for e in fp)
    visit_counts = Counter(visits).iteritems()
print '\n'.join(['{}\t{}'.format(ip, count) for (ip, count) in sorted(visit_counts, key=lambda (ip, count): count, reverse=True)[:5]])
```

```python
# 1. There is a set of IP addresses 10.140.[22-32].[10-181], find out the fastest server and print its address
# 2. improve the performance

from gevent import subprocess
import itertools
from gevent.pool import Pool

pool = Pool(100) # concurrent action count

ips = itertools.product((10, ), (140, ), range(22, 33), range(10, 182))

def get_response_time(ip):
    try:
        out = subprocess.check_output('ping -c 1 -W 1 {}.{}.{}.{}'.format(*ip).split())
        for line in out.splitlines():
            if 'time=' in line:
                return (ip, float(line.split('time=')[1].split()[0]))
    except subprocess.CalledProcessError:
        pass

    return (ip, None)

resps = pool.map(get_response_time, ips)
reachable_resps = filter(lambda (ip, resp): resp != None, resps)

print min(reachable_resps, key=lambda (ip, resp): resp)
```
